# OCPM Readout Generator

## Introduction

This notebook generates readouts of object / event metadata, attributes, transformations and relationships within an OCPM workspace. One excel report is generated, which includes five sheets:
- Object Attributes: Attribute data per object
- Object Relationships: Relationship data between objects
- Event Attributes: Attribute data per event
- Event Relationships: Relationship data between events and objects
- Transformations: Transformation data per object and event
- Attributes + Transformations: Transformation data per attribute per object / event (metadata of attribute and SQL snippet of where attribute is used also included)
